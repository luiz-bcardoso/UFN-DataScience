### Exercício Avaliativo Final – Análise de Dados de Astronautas

A agência espacial fictícia UFN CosmosLab precisa analisar os perfis dos astronautas que
participaram do seu mais recente programa de treinamento para missões em
microgravidade. Você foi contratado como analista de dados júnior para explorar as
informações coletadas e entregar insights iniciais à equipe de planejamento.

Você recebeu um arquivo CSV contendo os dados de 20 astronautas, incluindo nome,
idade, especialidade, tempo em gravidade zero, planeta favorito, e nota na avaliação do
treinamento.

--- 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/RobertsonWeb/materiais-numpy-pandas/refs/heads/main/datasets/astronautas_estacao_espacial.csv', sep=',')
df
```

1. Quantos astronautas têm idade registrada e quantos não têm?

```python
filtro_idade_nula = df['Idade'].isna()
print(f"Qunatidade de astronautas com idade não nula: {(~filtro_idade_nula).sum()}")
print(f"Quantidade de astronautas com idade nula: {filtro_idade_nula.sum()}")
```

2. Qual especialidade aparece com mais frequência na tripulação?

```python
filtro_maior_especialidade = df['Especialidade'].mode()[0]
print(f"Maior especialidade: {filtro_maior_especialidade}")
```

3. Qual é o planeta favorito mais comum?

```python
filtro_planeta = df['PlanetaFavorito'].max()
print(f"Planeta mais comum: {filtro_planeta}")
```

4. Entre os astronautas com mais de 500 horas em gravidade zero, qual foi a média da
avaliação no treinamento?

```python
filtro_astronautas_avaliacao = df['HorasEmGravidadeZero'] > 500
media_avaliacao_filto = df[filtro_astronautas_avaliacao]['AvaliaçãoTreinamento'].mean()
print(f"Média de avaliação dos astronautas com mais de 500 horas em gravidade zero: {media_avaliacao_filto:.2f}")
df[filtro_astronautas_avaliacao]
```
